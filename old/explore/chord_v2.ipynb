```python
import pandas as pd
import numpy as np
import holoviews as hv
from holoviews import opts, dim
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

hv.extension('bokeh')
hv.output(size=200)
```

```python
matrix = pd.read_csv('../data/pearson_matrix_2024-01-01_2025-01-01_top20.csv').set_index('Ticker')
matrix
```

```python
# Create a figure with a larger size
plt.figure(figsize=(120, 100))

# Create heatmap
sns.heatmap(matrix, 
            annot=False,  # Show correlation values
            cmap='RdBu',  # Red-Blue diverging colormap
            vmin=-1, vmax=1,  # Fix scale from -1 to 1
            center=0,  # Center colormap at 0
            fmt='.2f',  # Show 2 decimal places
            square=True,  # Make cells square
            xticklabels=matrix.columns,  # Show x-axis labels
            yticklabels=matrix.columns
            )  # Show y-axis labels

# Customize the plot
plt.title('Pearson Correlation Heatmap of Stock Returns', pad=20)
plt.xlabel('Stock Symbols')
plt.ylabel('Stock Symbols')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

```python
upper_triangle = np.triu(matrix, k=1)
num_edges = []
for r in np.linspace(0, 1, 1000):
    row_indices, col_indices = np.where(upper_triangle > r)
    num_edges.append(len(row_indices))
px.line(x=np.linspace(0, 1, 1000), y=num_edges, title='Number of Edges vs Threshold')
```

```python
upper_triangle = np.triu(matrix, k=1)
row_indices, col_indices = np.where(upper_triangle > 0.6)
print(len(row_indices))
edges = pd.DataFrame({
    'source': matrix.columns[col_indices],
    'target': matrix.columns[row_indices],
    'value': upper_triangle[row_indices, col_indices]
})
edges
```

```python
top_20 = pd.read_csv('../data/top_20_companies_by_sector.csv')
top_20_labels = top_20[['Symbol', 'Sector']].assign(Symbol_id=pd.factorize(top_20['Symbol'])[0]).assign(Sector_id=pd.factorize(top_20['Sector'])[0])
top_20_labels
```

```python
node_type = 'Sector'
group_type = 'Symbol'
mapped_edges = (edges
                .merge(top_20_labels, left_on='source', right_on='Symbol', how='left', suffixes=('_source', '_target'))
                .merge(top_20_labels, left_on='target', right_on='Symbol', how='left', suffixes=('_source', '_target'))
                .drop(columns=[ 'source', 'target', f'{group_type}_source', f'{group_type}_target', f'{node_type}_source', f'{node_type}_target', f'{group_type}_id_source', f'{group_type}_id_target'])
                .rename(columns={f'{node_type}_id_source': 'source', f'{node_type}_id_target': 'target'})
                )
mapped_edges
```

```python
mapped_nodes = (top_20_labels
                # .set_index(f'{node_type}_id')
                .drop(columns=[f'{group_type}_id'])
                .rename(columns={node_type: 'name', group_type: 'group', f'{node_type}_id': 'index'})
                )
mapped_nodes = (mapped_nodes[mapped_nodes['index'].isin(mapped_edges['source'].unique())]
                .drop_duplicates(subset=['name', 'index'])
                .set_index('index')
                )
mapped_nodes = hv.Dataset(pd.DataFrame(mapped_nodes), 'index')
mapped_nodes.data
# mapped_nodes
```

```python
chord = hv.Chord((mapped_edges, mapped_nodes))
# chord.opts(
#     opts.Chord(cmap='Category20', edge_cmap='Category20', edge_color=dim('source').str(), 
#                labels='group', node_color=dim('index').str()))
chord
```
