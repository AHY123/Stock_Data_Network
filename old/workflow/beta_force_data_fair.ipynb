```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
matrix = pd.read_csv('../data/beta_matrix_2024-01-01_2025-01-01_top50.csv').set_index('Ticker')
matrix
```

```python
labels = pd.read_csv('../data/top_50_companies_by_sector.csv')
labels
```

```python
upper_triangle = np.triu(matrix, k=1)
row_indices, col_indices = np.where(np.abs(upper_triangle) > 0)
print(len(row_indices))
graph = pd.DataFrame({
    'source': matrix.columns[col_indices],
    'target': matrix.columns[row_indices],
    'value': upper_triangle[row_indices, col_indices],
    'abs': np.abs(upper_triangle[row_indices, col_indices])
})
graph
```

```python
node_degrees = pd.concat([graph['source'], graph['target']]).value_counts()

def filter_top_edges_for_node(df, node_degrees, min_edges=10):
    result_edges = []
    
    for node in node_degrees.index:
        # Get all edges for this node (both as source and target)
        node_edges = df[(df['source'] == node) | (df['target'] == node)]
        
        # Sort edges by absolute value to get strongest connections
        node_edges = node_edges.sort_values(by='value', key=abs, ascending=False)
        
        # If node has fewer than min_edges, keep all edges
        if len(node_edges) < min_edges:
            result_edges.append(node_edges)
        else:
            # Keep top min_edges edges
            result_edges.append(node_edges.head(min_edges))
    
    # Combine all edges and remove duplicates
    filtered_df = pd.concat(result_edges).drop_duplicates().reset_index(drop=True)
    
    # Verify each node has at least min_edges
    for node in node_degrees.index:
        node_edges = filtered_df[(filtered_df['source'] == node) | (filtered_df['target'] == node)]
        if len(node_edges) < min_edges:
            print(f"Warning: Node {node} still has only {len(node_edges)} edges")
    
    return filtered_df

links = filter_top_edges_for_node(graph, node_degrees, 2)
links
```

```python
relevant_nodes = set(links['source'].unique()) | set(links['target'].unique())
nodes = (labels[labels['Symbol']
        .isin(relevant_nodes)]
        [['Symbol', 'Sector', 'Market Cap (B)', 'Last Sale']]
        .rename(columns={
            'Symbol': 'id',
            'Sector': 'group',
            'Market Cap (B)': 'marketCap',
            'Last Sale': 'price'
            })
        )
nodes
```
